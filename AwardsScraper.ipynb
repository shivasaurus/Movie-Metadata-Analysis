```python
from bs4 import BeautifulSoup
import csv
import requests
import numpy as np
```

```python
url = "https://www.oscars.org/oscars/ceremonies/"
dictList = []
for i in range(1990, 2018):
    url1 = url+str(i) 
    results = requests.get(url1)
    page = results.content
    soup = BeautifulSoup(page)
    divList = soup.find_all("div", "view-grouping-header")
    nomineeDict = {}
    for i in range(0, len(divList)):
        div = divList[i]
        categoryNominees = []
        category = div.find("h2")
        if category == None:
            break
        header = div.find_next_sibling("div", "view-grouping-content")

        if (i < 4):
            nominees = header.find_all("div", "views-field-title")
            for nominee in nominees:
                categoryNominees.append(nominee.find("span").get_text().strip())
                #print(nominee.find("span").get_text().strip())
        else:
            nominees = header.find_all("div", "views-field-field-actor-name")
            for nominee in nominees:
                categoryNominees.append(nominee.find("h4").get_text())
                #print(nominee.find("h4").get_text())
        #print(category.get_text())
        nomineeDict[category.get_text()] = categoryNominees
    print(nomineeDict)
    dictList.append(nomineeDict)
```

```python
with open('oscarNoms.csv', 'w', encoding='utf8', newline='') as output_file:
    fc = csv.DictWriter(output_file, 
                        fieldnames=set().union(*(d.keys() for d in dictList)))
    fc.writeheader()
    fc.writerows(dictList)
```
